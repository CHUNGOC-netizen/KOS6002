```python
data3 = np.random.normal(loc=-5, scale=2, size=249)
data4 = np.random.laplace(loc=0, scale=1, size=10000)
data5 = np.random.exponential(scale=1, size=1000)
```

```python
datalist = [data1, data2, data3, data4, data5]
plt.violinplot(datalist, showmeans=True, showextrema=True)
for i, data in enumerate(datalist):
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    plt.plot([i + 1, i + 1], [q1, q3], color='red', linewidth=3, label='IQR' if i == 0 else "")
if len(datalist) > 0:
    plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.violinplot(datalist, showmeans=True, showextrema=True, showmedians=False)
for idx, data in enumerate(datalist):
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    plt.plot([idx + 1, idx + 1], [q1, q3], color='red', linewidth=3, label='IQR' if idx == 0 else "")
    # Outliers: points outside 1.5*IQR
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    outliers = data[(data < lower_bound) | (data > upper_bound)]
    plt.scatter([idx + 1] * len(outliers), outliers, color='black', s=10, label='Outliers' if idx == 0 else "")
if len(datalist) > 0:
    plt.legend()
plt.grid(True)
plt.show()
```
