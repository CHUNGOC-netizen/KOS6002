```python
import random

numbers = []

for i in range(12):
    r = random.randint(-50, 50)
    numbers.append(r)

print(numbers)

import matplotlib.pyplot as plt

plt.plot(numbers, marker='o')
plt.show()
```

```python
# 평균값 계산하기
총합 = numbers[0] + numbers[1] + numbers[2] + numbers[3] + numbers[4] + numbers[5] + numbers[6] + numbers[7] + numbers[8] + numbers[9] + numbers[10] + numbers[11]
평균값 = 총합 / len(numbers)
print("평균값:", 평균값)
```

```python
total = 0
for num in numbers:
    total += num
average = total / len(numbers)

print("총합:", total)
print("평균값:", average)
```

정규분포 = normal distribution = Gaussian distribution

```python
standard_normal_samples = np.random.randn(1000)

plt.plot(standard_normal_samples, marker='o', markersize=3)
```

```python
# histogram
plt.hist(standard_normal_samples, bins=10, edgecolor='black');
```

```python
standard_normal_samples.std() # 표준편차 = standard deviation
```

```python
samples_10000 = np.random.randn(10000)
print(samples_10000)
```

```python
samples_10000_normal = np.random.normal(loc=0, scale=1, size=10000)
plt.hist(samples_10000_normal, bins=30, edgecolor='black');
```

```python
samples_10000_normal_loc4 = np.random.normal(loc=4, scale=1, size=10000)
plt.hist(samples_10000_normal_loc4, bins=30, edgecolor='black');
```

```python
samples_10000_normal_scale2 = np.random.normal(loc=0, scale=2, size=10000)
plt.hist(samples_10000_normal_scale2, bins=30, edgecolor='black');
```

```python
plt.hist(samples_10000_normal, bins=30, alpha=0.5, label='N(0,1)')
plt.hist(samples_10000_normal_loc4, bins=30, alpha=0.5, label='N(4,1)')
plt.hist(samples_10000_normal_scale2, bins=30, alpha=0.5, label='N(0,2)')
plt.legend()
plt.show()
```
